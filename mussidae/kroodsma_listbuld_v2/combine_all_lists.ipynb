```python
import pandas as pd
import json
import csv
```

```python
def isNaN(num):
    return num != num
```

```python
directories = ['input_lists','input_lists_reviewed']
```

```python
from os import listdir
from os.path import isfile, join
mypath = directories[0]
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) and ".csv" in f]
```

```python
rows = []

for d in directories:
    mypath = d
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f)) and ".csv" in f]

    for f in onlyfiles:
        print f
        with open(mypath+"/"+f.replace('.csv','.json')) as data_file:    
            data = json.load(data_file)
        df = pd.read_csv(mypath+"/"+f) 
    #     print df.head()
        df = df.set_index([data['headers']['mmsi']])
        for index, row in df.iterrows():

            label = row[data['headers']['label']]
                    
            if data['mappings']:
                if label in data['mappings']:
                    label = data['mappings'][label]
                    if label:
                        label = label.lower()
                        
            if label and not isNaN(label): # annoying nans
                label = label.lower()

            if data['headers']['length']:
                length = row[data['headers']['length']]
            else: 
                length = None  

            if data['headers']['tonnage']:
                tonnage = row[data['headers']['tonnage']]
            else: 
                tonnage = None  

            if data['headers']['engine power']:
                engine_power = row[data['headers']['engine power']]
            else: 
                engine_power = None  

            source = f.replace(".csv","")
            
            if label is not None or length is not None or tonnage is not None or engine_power is not None:
                rows.append([index,label,length,tonnage,engine_power,source])

    #     print df.head()   
    
    

```

```python
from operator import itemgetter
rows = sorted(rows, key=itemgetter(0))
```

```python
r = rows[0]
[new_mmsi, label, length, tonnage,engine_power, source] = r

new_rows = []

mmsi = r[0]
new_row = [mmsi]+[[r[i]] for i in range(1,len(r))]

for i in range(1,len(rows)):
    r = rows[i]
    if r[0] == mmsi:
        for i in range(1,6):
            if(r[i]):
                new_row[i].append(r[i])
    else:  
        new_rows.append(new_row)
        new_row = [r[0]]+[[r[i]] for i in range(1,len(r))]
        mmsi = r[0]
```
