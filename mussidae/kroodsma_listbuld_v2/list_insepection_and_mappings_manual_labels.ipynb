```python
import csv
import pandas as pd
import numpy as np
headers = {}
label_mappings = {}
counts = {}
import json
```

```python
def save_file(filename):
    js = {
        'mappings': label_mappings[filename],
        'counts': counts[filename],
        'headers': headers[filename] 
    }

    js_towrite = json.dumps(js)
    with open('input_lists_reviewed/'+filename+'.json','w') as f:
        f.write(js_towrite)

def get_counts(label_header,filename,df):
    c={}
    for a in df.groupby(label_header).indices:
        c[a]= len(df.groupby(label_header).indices[a])
    return c
```

```python
filename = 'Maybe Reefers'
df = pd.read_csv("input_lists_reviewed/"+filename+".csv")
```

```python
df.head()
```

```python
headers[filename] = { 'mmsi':'mmsi',
                                    'label':'Vessel Identity',
                                    'tonnage':None, 
                                    'length':None, 
                                    'engine power': None}
```

```python
counts[filename] = get_counts(headers[filename]['label'],filename,df)
counts[filename]
```

```python
label_mappings[filename] = None
```

```python
save_file(filename)
```
